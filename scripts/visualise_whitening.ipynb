```python
path = 'C:\\Users\\ConnardMcGregoire\\elk-reporters\\gpt2\\imdb\\'
burns = 'burnsnorm\\eval.csv'
cluster_no_whitening = 'cluster_no_whitening\\eval.csv'
cluster_whitening = 'cluster_whitening\\eval.csv'
```

```python
# import libraries to read csv files
import pandas as pd
```

```python
burns = pd.read_csv(path+burns)
cluster_no_whitening = pd.read_csv(path+cluster_no_whitening)
cluster_whitening = pd.read_csv(path+cluster_whitening)

burns_acc = burns['cal_acc_upper']
cluster_no_whitening_acc = cluster_no_whitening['cal_acc_upper']
cluster_whitening_acc = cluster_whitening['cal_acc_upper']
```

```python
import matplotlib.pyplot as plt

plt.plot(burns_acc, label='burns')
plt.plot(cluster_no_whitening_acc, label='cluster')
plt.plot(cluster_whitening_acc, label='cluster_whitening')
plt.xlabel('layer * 3')
plt.ylabel('cal_acc_upper')
plt.legend()
plt.show()
```
